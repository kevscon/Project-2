```python
poly_pipe = pipeline.Pipeline([
#     ("std", preprocessing.StandardScaler()),
    ("poly", preprocessing.PolynomialFeatures(1)),
    ("ols", linear_model.LinearRegression())
])
poly_pipe.get_params()
```

```python
poly_pipe.set_params(poly__degree=2, poly__interaction_only = True) 

# Fit the model against the training data

poly_pipe.fit(xtrain, ytrain)
# Evaluate the model against the testing data
print(
    f"train r2: {poly_pipe.score(xtrain, ytrain):.2}, "
    f"test r2: {poly_pipe.score(xtest,ytest):.2}")
```

```python
# higher degree models do not improve prediction!
# plot test/train curve
figsize(10,5)
train_scores = []
test_scores = []
degrees = range(1,4,1)
for i,degree in enumerate(degrees):
    poly_pipe.set_params(poly__degree=degree) 
    poly_pipe.fit(xtrain, ytrain)
    train_scores.append(poly_pipe.score(xtrain, ytrain))
    test_scores.append(poly_pipe.score(xtest,ytest))

plot(degrees, 1-np.array(train_scores), label="train")
plot(degrees, 1-np.maximum(test_scores,0), label="test")
xlabel("degree of polynomial transformation")
ylabel("$1-r^2$ w floor at 0")
title("learning curve")
legend();
```
