## Challenge 1

Plot domestic total gross over time.

```python
df_movies_clean = df_movies
```

```python
df_movies_clean.head()
```

```python
# define function to re-format monetary values to integers
def money_to_int(moneystring):
    
    try:
        moneystring = moneystring.replace('$', '').replace(',', '')
        return int(moneystring)
    
    except ValueError:
        return None
```

```python
# apply function to dom_gross column
df_movies_clean['dom_gross'] = df_movies_clean['dom_gross'].apply(money_to_int)
```

```python
# remove rows with NaN values
df_movies_clean['dom_gross'] = df_movies_clean['dom_gross'].dropna(how='any')
```

```python
# define function to re-format date values
import dateutil.parser

def to_date(datestring):
    
    try:
        date = dateutil.parser.parse(datestring)
        return date
    
    except ValueError:
        return None
```

```python
# apply function to release_date colum
df_movies_clean['release_date'] = df_movies_clean['release_date'].apply(to_date)
```

```python
# remove rows with NaN values
df_movies_clean['release_date'] = df_movies_clean['release_date'].dropna(how='any')
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as dates

x = df_movies_clean['release_date']
y = df_movies_clean['dom_gross']
plt.scatter(x.values, y);
```

## Challenge 2

Plot runtime vs. domestic total gross.

```python
# define function to re-format runtime
def runtime_to_minutes(runtimestring):
    runtime1 = str(runtimestring)
    runtime = runtime1.split()
    try:
        minutes = int(runtime[0])*60 + int(runtime[2])
        return minutes
    except:
        return None
```

```python
# apply function to runtime column
df_movies_clean['runtime'] = df_movies_clean['runtime'].apply(runtime_to_minutes)
```

```python
x = df_movies_clean['runtime']
y = df_movies_clean['dom_gross']
plt.scatter(x, y);
```

```python
df_movies_clean['year'] = df_movies_clean['release_date'].map(lambda x: x.year)
df_movies_clean.head()
```

## Challenge 5

What director in your dataset has the highest gross per movie?

```python
df_direct = df_movies_clean.dropna()
```

```python
df_direct['director'] = df_direct['director'].dropna(how='any')
```

```python
df_direct
```

```python
a = df_direct.groupby('director')['dom_gross'].sum()
b = df_direct.groupby('director')['dom_gross'].count()
```

```python
# James Cameron has the highest gross per movie in this dataset
c = a/b
c.sort_values(ascending=False).head()
```
