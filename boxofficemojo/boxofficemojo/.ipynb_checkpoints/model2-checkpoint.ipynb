```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.formula.api as smf
import patsy
import pickle

import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import RidgeCV
%matplotlib inline
%pylab

from sklearn import preprocessing, linear_model, pipeline
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('data_1.csv')
```

## Model

Create design matricies
Test/train split
First model

```python
# set up matrices

y, X = patsy.dmatrices('intl_gross ~ budget + runtime + num_theaters + num_days + genre \
                        + mpaa_rating + release_month + release_year', data=df, return_type="dataframe")
```

```python
# split data into test and train sets
xtrain, xtest, ytrain, ytest = train_test_split(X, y)
```

```python
# initial model based on training data
model = LinearRegression()
model.fit(xtrain, ytrain)
model.score(xtest, ytest)
```

```python
# benchmark for budget vs intl_gross
# very low, difficult dataset - high volatiltiy in the market?
model_b = linear_model.LinearRegression()
model_b.fit(xtrain[['budget']], ytrain)
model_b.score(xtest[['budget']], ytest)
```

```python
X.columns
```

```python
# plot coefficients - Action/Fantasy genre
figsize(20, 10)

scatter(np.arange(len(X.columns)), np.abs(model.coef_))
plt.xticks(np.arange(len(model.coef_)), X.columns, rotation=90);
plt.title('Model Coefficients', fontsize=15)
plt.xlabel('Features', fontsize=14)
plt.ylabel('Domestic Gross Total', fontsize=12);
```

```python
y_predict = model.predict(xtest)
y_predict
```

```python
# plot prediction on test data for budget
# had limited budget data, so filled NaNs with mean values
plt.scatter(xtest['budget'], ytest['intl_gross']);
plt.title('Model Prediction based on Budget');
plt.xlabel('Movie Budget');
plt.ylabel('Internation Gross');
plt.scatter(xtest['budget'], y_predict, c='red', marker='x');
```

```python
# try higher degree model
poly_pipe = pipeline.Pipeline([
#     ("std", preprocessing.StandardScaler()),
    ("poly", preprocessing.PolynomialFeatures(1)),
    ("ols", linear_model.LinearRegression())
])
poly_pipe.get_params()
```

```python
# higher degree model increases training fit but is very poor for the test
poly_pipe.set_params(poly__degree=2, poly__interaction_only = True) 

# Fit the model against the training data

poly_pipe.fit(xtrain, ytrain)
# Evaluate the model against the testing data
print(
    f"train r2: {poly_pipe.score(xtrain, ytrain):.2}, "
    f"test r2: {poly_pipe.score(xtest,ytest):.2}")
```

```python
# higher degree models do not improve prediction!
# plot test/train curve
figsize(10,5)
train_scores = []
test_scores = []
degrees = range(0,4,1)
for i,degree in enumerate(degrees):
    poly_pipe.set_params(poly__degree=degree) 
    poly_pipe.fit(xtrain, ytrain)
    train_scores.append(poly_pipe.score(xtrain, ytrain))
    test_scores.append(poly_pipe.score(xtest,ytest))

plot(degrees, 1-np.array(train_scores), label="train")
plot(degrees, 1-np.maximum(test_scores,0), label="test")
xlabel("degree of polynomial transformation")
ylabel("$1-r^2$ w floor at 0")
title("learning curve")
legend();
```
